# Deal conversion rate

```python
# Calculating deal conversion rate
total_pitches = df['pitch_number'].nunique()
deals = df[df['deal'] == 1]['pitch_number'].nunique()
deal_conversion_rate = (deals / total_pitches) * 100


# Plotting
labels = ['Deals', 'No Deals']
sizes = [deals, total_pitches - deals]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Overall Deal Conversion Rate')
plt.show()

print(f"Overall Deal Conversion Rate: {deal_conversion_rate:.2f}%")
```

**Insight:**
The participants have more than 50% chance that they get a deal .

# Comparison on Total Investment and Number of deals by each Sharks 

```python

sharks = ["ashneer_deal", "anupam_deal", "aman_deal", "namita_deal", "vineeta_deal", "peyush_deal", "ghazal_deal"]

# Calculating the number of sharks involved in each deal
df['num_sharks'] = df[sharks].sum(axis=1)

# Group by number of sharks and calculate the total investment for each group
investment_summary = df.groupby('num_sharks')['deal_amount'].sum().reset_index()

# Filtering out rows where num_sharks is 0 (deals with no sharks involved)
investment_summary = investment_summary[investment_summary['num_sharks'] > 0]

# Plotting the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(investment_summary['num_sharks'], investment_summary['deal_amount'], color='skyblue')

# Title and labels
plt.title('Total Investment by Number of Sharks Involved')
plt.xlabel('Number of Sharks Involved')
plt.ylabel('Total Investment Amount (USD)')
plt.xticks(investment_summary['num_sharks'])
plt.grid(axis='y')

# Adding value labels inside the bars, rotated counterclockwise
for bar in bars:
    yval = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # x-position of the text
        yval - (yval * 0.05),  # y-position slightly inside the bar
        f'{yval:.2f}',  # Formatting the text to 2 decimal places
        ha='center',  # Center align the text horizontally
        va='top',  # Position the text at the top of the bar
        fontsize=12,  # Font size
        #rotation=90,  # Rotate counterclockwise 90 degrees
        color='Darkblue',  # Text color
        fontweight='bold'  # Bold font for better visibility
    )

plt.tight_layout()
plt.show()
```

**Insights:**
Higher total investments are made when 1-3 sharks are involved, while investments decrease significantly with 4-5 sharks.

# Comparision between Total amount invested and number of sharks involved

**Insights:**


1. Aman and Peyush are the most active sharks both in terms of the number of deals and investment amounts.

# Pitches made and Deal closed per Episode 

```python


# Calculating the number of pitches made per episode
pitches_made_per_episode = df.groupby("episode_number")["pitch_number"].count()

# Calculating the number of deals closed per episode
deals_closed_per_episode = df[df["deal"] > 0].groupby("episode_number")["deal_amount"].count()

# Creating a summary DataFrame to hold pitches and deals per episode
summary_df = pd.DataFrame({
    "Pitches Made": pitches_made_per_episode,
    "Deals Closed": deals_closed_per_episode,
}).reset_index()

# Filling NaN values with 0 in case there are episodes with pitches but no deals
summary_df = summary_df.fillna(0)

# Plotting the line plot
plt.figure(figsize=(12, 6))

# Plot for Pitches Made
plt.plot(summary_df['episode_number'], summary_df['Pitches Made'], marker='o', linestyle='-', color='green', label='Pitches Made')

# Plot for Deals Closed
plt.plot(summary_df['episode_number'], summary_df['Deals Closed'], marker='o', linestyle='-', color='orange', label='Deals Closed')

# Titles and labels
plt.title('Number of Pitches Made and Deals Closed per Episode')
plt.xlabel('Episode Number')
plt.ylabel('Count')
plt.xticks(summary_df['episode_number'])
plt.grid(axis='y')

plt.legend()
plt.tight_layout()
plt.show()
```

**Insights:**


1. The number of pitches made remains consistently around 3 across most episodes.  

2. The number of deals closed fluctuates significantly, ranging from 0 to 3, with some episodes showing no deals despite multiple pitches.



3. Episodes 16 and 29 highlight a high number of pitches but low deal success, indicating challenging negotiations or less favorable outcomes.

# Shark Participation in Pitches, Number of Deals, and Acceptance Rate

```python
sharks = ['ashneer_present', 'anupam_present', 'aman_present', 'namita_present', 'vineeta_present', 'peyush_present' ]

# Shark Participation in Pitches
shark_participation = df[sharks].sum()

# Number of Deals by Each Shark
deals_columns = [col.replace('_present', '_deal') for col in sharks]
shark_deals = df[deals_columns].sum()

# Acceptance Rates
acceptance_rates = []
for participation, deals in zip(shark_participation, shark_deals):
    if participation > 0:
        acceptance_rates.append((deals / participation) * 100)
    else:
        acceptance_rates.append(0)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(12, 7))
x = range(len(sharks))

# Bar width and positions for each group
bar_width = 0.25
bar1 = [i - bar_width for i in x]  # For Participation in Pitches
bar2 = [i for i in x]              # For Number of Deals
bar3 = [i + bar_width for i in x]  # For Acceptance Rate

# Plotting bars for each category
plt.bar(bar1, shark_participation, width=bar_width, label='Participation in Pitches', color='skyblue')
plt.bar(bar2, shark_deals, width=bar_width, label='Number of Deals', color='salmon')
plt.bar(bar3, acceptance_rates, width=bar_width, label='Acceptance Rate (%)', color='lightgreen')

# Adding data labels on top of the bars using .iloc to handle Series indexing
for i in range(len(sharks)):
    plt.text(bar1[i], shark_participation.iloc[i] + 1, str(shark_participation.iloc[i]), ha='center', va='bottom')
    plt.text(bar2[i], shark_deals.iloc[i] + 1, str(shark_deals.iloc[i]), ha='center', va='bottom')
    plt.text(bar3[i], acceptance_rates[i] + 1, f'{acceptance_rates[i]:.1f}%', ha='center', va='bottom')

# Adding gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels and title
plt.xlabel('Sharks')
plt.ylabel('Number of Pitches/Deals & Acceptance Rate (%)')
plt.title('Shark Participation in Pitches, Number of Deals, and Acceptance Rate')
plt.xticks(x, ['Ashneer', 'Anupam', 'Aman', 'Namita', 'Vineeta', 'Peyush'])
plt.legend()

# Overall summary statistics
total_participation = shark_participation.sum()
total_deals = shark_deals.sum()
plt.text(0.5, max(shark_participation.max(), shark_deals.max(), max(acceptance_rates)) * 1.15,
         f'Total Pitches: {total_participation}\nTotal Deals: {total_deals}',
         ha='center', va='bottom', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

**Insights:**

1 . Shark Anupam was present in most number of pitches , but has the lowest acceptance rate .

2 . Shark Peyush has the greatest acceptance rate .

# Total Investment and Success rate by Industry 

```python
# Calculating Total Investment per Industry
investment_by_industry = df.groupby('Industry')['deal_amount'].sum().reset_index()
investment_by_industry = investment_by_industry.sort_values(by='deal_amount', ascending=False)

# Calculating Success Rate per Industry
deals_by_industry = df.groupby('Industry')['deal'].apply(lambda x: (x > 0).sum()).reset_index()
deals_by_industry.columns = ['Industry', 'Deals Closed']
deals_by_industry['Total Pitches'] = df['Industry'].value_counts().reset_index(drop=True)
deals_by_industry['Success Rate'] = (deals_by_industry['Deals Closed'] / deals_by_industry['Total Pitches']) * 100

# Merging both dataframes to align industries
merged_df = pd.merge(investment_by_industry, deals_by_industry, on='Industry')

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

# Horizontal bar chart for Total Investment
ax1.barh(merged_df['Industry'], merged_df['deal_amount'], color='skyblue', label='Total Investment')
ax1.set_xlabel('Total Investment', fontsize=12)
ax1.set_ylabel('Industry', fontsize=12)

# Creating a second axis for Success Rate
ax2 = ax1.twiny()  # Create a twin of the horizontal axis
ax2.plot(merged_df['Success Rate'], merged_df['Industry'], 'o-', color='salmon', label='Success Rate (%)', linewidth=3)
ax2.set_xlabel('Success Rate (%)', fontsize=12)

# Adding titles and labels
plt.title('Total Investment vs. Success Rate by Industry', fontsize=16)
ax1.legend(loc='upper right')
ax2.legend(loc='lower right')

plt.show()
```

**Insights**

Food & Beverages , Healtcare & Wellness and Technology & Innovation were the most popular categories regarding Investments.

However, Technology & Investment and Transportation & Mobility had more success rates in getting the deal .
